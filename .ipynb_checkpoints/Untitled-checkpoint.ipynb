```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mysql.connector 
```

```python
from dotenv import load_dotenv
import os
load_dotenv()
db=mysql.connector.connect(host="localhost",username='root',password=os.getenv('PASSWORD'),database='ecommerce')
```

```python
cur=db.cursor()
```

# List all unique cities where customers are located.

```python
query=""" select distinct customer_city from customers """
cur.execute(query)
data=cur.fetchall()
data
```

# Count the number of orders placed in 2017.

```python
query=""" SELECT count(*) FROM orders where year(order_purchase_timestamp)=2017 """
cur.execute(query)
data=cur.fetchall()
data[0][0]
```

# Find the total sales per category.

```python
query=""" SELECT sum(price),product_id FROM order_items group by product_id; """
cur.execute(query)
data=cur.fetchall()
data
```
